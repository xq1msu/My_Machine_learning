Thus all the data management is done, to build the model, first we need to build a linear regression classifier and the sigmoid function

```python
def linear_Classifier(X_i, grad):
    # this function return a value to show if the points are above or below the plane(w_i, w_0)
    w_i = grad[0]
    w_0 = grad[1]
    return np.dot(w_i.T, X_i) + w_0 

def sigmoid(y_pre):
    # this function takes the linear classifier value, to show the possibility if it's Class A or Class B
    # in other worlds, we made the linear_classifier value to 1 or 0 now, to fit two class
    # this is the activation function
    return 1 / (1 + np.exp(-y_pre))
```

This notebook is used to make a Logistic Regression instance.

First of all, import all the necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.model_selection import train_test_split
```

### Parameters

```python
num_features = 4

learning_rate = 0.01
num_epochs = 400
batch_size = 10
```

### Step 1: Preparing Data


```python
df = pd.read_csv("Iris.csv")
 
df["Species"] = pd.Categorical(df["Species"]).codes
df = df.drop("Id", axis=1)
df = df[(df["Species"] == 1) | (df["Species"] == 0)]

df.head()
```

```python
df
```

### Step 2 : Seperate Data to X and y

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("Species", axis=1), df["Species"], test_size=0.25)
```

### Step 3 : convert [0,255] to [0,1]

```python
X_train, X_test = np.array(X_train, np.float32), np.array(X_test, np.float32)
y_train, y_test = np.array(y_train, np.float32), np.array(y_test, np.float32)
```

### Step 4 : Suffling Data and batch then by using tensorflow

```python
def load_array(data_arrays, batch_size):
    return tf.data.Dataset.from_tensor_slices(data_arrays).shuffle(buffer_size=1000).batch(batch_size)

data_iter = load_array((X_train, y_train), batch_size)
```

### Step 5 : Defining the module and Loss function

```python
# setting the module and Activation function
net = tf.keras.Sequential()
net.add(tf.keras.layers.Dense(1, input_shape=(num_features,), activation='sigmoid'))


# Setting Loss function
loss = tf.keras.losses.BinaryCrossentropy()
```

```python
net.compile(optimizer=tf.keras.optimizers.SGD(learning_rate = learning_rate),
           loss='binary_crossentropy',
           metrics=['acc'])
```

```python
history = net.fit(data_iter,epochs=num_epochs)
```

```python
acc = np.sum(tf.cast(tf.equal(np.around(net(X_test)).T, y_test), tf.float32))/len(y_test) * 100
print("This modle's accuracy is {} %".format(acc))
```
